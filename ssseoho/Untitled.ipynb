```python
#-*- coding: utf-8 -*-
data1 = pd.read_csv("./health_data/2016_health_checkup_1.csv")
data2 = pd.read_csv("./health_data/2016_health_checkup_2.csv") 
data3 = pd.read_csv("./health_data/2016_health_checkup_3.csv") 
data4 = pd.read_csv("./health_data/2016_health_checkup_4.csv") 
data = data1.append(data2).append(data3).append(data4)
data = data.reset_index(drop=True)
data
```

결측치가 다음과 같이 하나만 존재한 경우 모두 정상치에 해당됨. 두개 다 결측치가 있는 경우에는 위의 두 값을 제외하고는 다 동일한 데이터이다.(87개의 동일한 데이터)

```python
normal = (data['수축기혈압'] < 120) & (data['이완기혈압'] < 80)
data[normal][['수축기혈압','이완기혈압']]
```

정상치에 대한 데이터들의 평균값을 구해 대입을 하기로 함.(수축기혈압 결측 & 이완기혈압 결측)

```python
mean_contract = data[normal]['수축기혈압'].mean().round(0)
print("수축기혈압 평균 :",mean_contract)
mean_relax = data[normal]['이완기혈압'].mean().round(0)
print("이완기혈압 평균 :",mean_relax)
```
